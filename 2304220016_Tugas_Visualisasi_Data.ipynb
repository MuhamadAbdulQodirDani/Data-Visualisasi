` Kolom dalam Dataset`
* survived (int) : status keberlangsungan hidup (0 = Tidak Selamat & 1 = Selamat)
* pclass (int) : Kelas tiket (1 = First, 2 = Second, 3 = Third)

* age (float) : Usia penumpang
* sibsp (int) : Jumlah saudara kandung atau pasangan yang ada di kapal
* parch (int) : Jumlah orang tua atau anak di kapal

* who (str) : Kategori orang (man, woman, child)
* adult_male (bool) : Apakah penumpang adalah pria dewasa
* deck (str) : Dek tempat kabin penumpang berada
* embark_town (str) : Nama kota keberangkatan (Cherbourg, Queenstwon, Southampton)
* alive (str) : Status kelangsungan hidup dalam format teks (yes/no), kolom ini adalah representasi teks dari 'survived'
* alone (bool) : Apakah penumpang bepergian sendirian (True/False) 


### *Dataset Titanic*

```python
# Load Data titanic yang telah diprocessing sebelumnya
import pandas as pd
data = pd.read_excel('D:/SEMESTER 4/Penambangan Data dan Analisis Bisnis/Week 4/data_bersih.xlsx')
data
```

### *Visualisasi Data Titanic*

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Figure setup
# fig, axes = plt.subplots(2, 2, figsize=(15, 10))
```

```python
# 1. Distribusi Kelangsungan Hidup
sns.countplot(x="survived", data=data, palette="pastel")
plt.title("Distribusi Kelangsungan Hidup (0 = Tidak Selamat, 1 = Selamat)")
# Show plot
plt.tight_layout()
plt.show()
```

`interpretasi:`

Grafik tersebut menunjukkan distribusi kelangsungan hidup penumpang Titanic, di mana nilai 0 mewakili penumpang yang tidak selamat, dan nilai 1 mewakili mereka yang selamat. Dari grafik, terlihat bahwa jumlah penumpang yang tidak selamat jauh lebih banyak dibandingkan dengan yang selamat. Hal ini mencerminkan kenyataan bahwa mayoritas penumpang Titanic tidak berhasil bertahan dalam kecelakaan tragis tersebut. Perbedaan jumlah yang signifikan antara kedua kategori ini menunjukkan betapa besarnya dampak dari bencana tersebut terhadap keseluruhan penumpang kapal. Jika dianalisis lebih lanjut, faktor-faktor seperti kelas tiket, jenis kelamin, dan usia kemungkinan besar berperan dalam menentukan peluang kelangsungan hidup.

```python
# 2. Distribusi Usia Penumpang
sns.histplot(data["age"].dropna(), bins=30, kde=True, color="skyblue")
plt.title("Distribusi Usia Penumpang")
# Show plot
plt.tight_layout()
plt.show()
```

`interpretasi:`


Grafik ini menunjukkan distribusi usia penumpang Titanic. Sumbu horizontal (X) mewakili usia penumpang, sedangkan sumbu vertikal (Y) menunjukkan jumlah penumpang dalam setiap rentang usia. Dari histogram ini, terlihat bahwa mayoritas penumpang berusia antara 20 hingga 40 tahun, dengan puncak distribusi berada di sekitar usia 30 tahun dikarenakan beberapa data hilang yang telah terisi mean. Grafik juga menunjukkan bahwa ada sejumlah kecil penumpang yang berusia sangat muda (di bawah 10 tahun) maupun sangat tua (di atas 60 tahun). Pola distribusi ini mengindikasikan bahwa sebagian besar penumpang Titanic berada dalam rentang usia produktif, meskipun terdapat variasi usia yang cukup luas.

```python
# 3. Kelangsungan Hidup Berdasarkan Kelas Tiket
sns.countplot(x="pclass", hue="survived", data=data, palette="coolwarm")
plt.title("Kelangsungan Hidup Berdasarkan Kelas Tiket")
# Show plot
plt.tight_layout()
plt.show()
```

`interpretasi:`

Grafik ini menunjukkan hubungan antara kelas tiket penumpang Titanic (pclass) dan kelangsungan hidup mereka (survived), di mana 0 berarti tidak selamat dan 1 berarti selamat.

Dari grafik, terlihat bahwa:
1. Kelas 1 memiliki lebih banyak penumpang yang selamat dibandingkan yang tidak selamat, menunjukkan bahwa penumpang kelas atas memiliki peluang lebih besar untuk bertahan.
2. Kelas 2 memiliki jumlah penumpang selamat dan tidak selamat yang hampir seimbang.
3. Kelas 3 memiliki jumlah korban terbanyak, dengan jumlah penumpang yang tidak selamat jauh lebih tinggi dibandingkan yang selamat.

Kesimpulannya, kelas tiket memiliki pengaruh besar terhadap peluang selamat. Penumpang kelas 1 memiliki tingkat kelangsungan hidup tertinggi, sedangkan penumpang kelas 3 mengalami tingkat kematian tertinggi, mencerminkan ketimpangan sosial yang terjadi dalam penyelamatan saat tragedi Titanic.

```python
# 4. Kelangsungan Hidup Berdasarkan Jenis Kelamin
sns.countplot(x="sex", hue="survived", data=data, palette="viridis")
plt.title("Kelangsungan Hidup Berdasarkan Jenis Kelamin")
# Show plot
plt.tight_layout()
plt.show()
```

`interpretasi:`

Grafik tersebut menunjukkan hubungan antara jenis kelamin penumpang dan kelangsungan hidup mereka dalam tragedi Titanic. Dari grafik, terlihat bahwa mayoritas korban yang tidak selamat adalah laki-laki, dengan jumlah yang jauh lebih tinggi dibandingkan perempuan. Sebaliknya, sebagian besar perempuan berhasil selamat, menunjukkan bahwa mereka memiliki peluang bertahan lebih besar dibandingkan laki-laki. Jumlah laki-laki yang selamat jauh lebih sedikit dibandingkan yang tidak selamat, yang mengindikasikan bahwa mereka memiliki risiko kematian yang lebih tinggi. Pola ini mencerminkan kebijakan "women and children first" yang diterapkan dalam proses evakuasi, di mana perempuan dan anak-anak lebih diprioritaskan untuk mendapatkan tempat di sekoci penyelamat.
